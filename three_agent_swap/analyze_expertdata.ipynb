```python
import numpy as np
import matplotlib.pyplot as plt

# Replace these with your actual arrays of shape (100, 100, 2)
# For demonstration purposes, we're creating dummy arrays here:
arr1 = np.random.randn(100, 100, 2)
arr2 = np.random.randn(100, 100, 2)
arr3 = np.random.randn(100, 100, 2)

def plot_all(arr1, arr2, arr3):
    """
    Plots the first trajectory (index 0) from each of three (100,100,2) arrays.
    """
    
    plt.figure(figsize=(8, 6))
    plt.xlim(-3, 5)
    plt.ylim(-3, 3)
    for i in range(len(arr1)):
        t1 = arr1[i]
        t2 = arr2[i]
        t3 = arr3[i]
        plt.scatter(t1[0, 0], t1[0, 1], label='Array 1 Traj 0', color='blue')
        # plt.plot(t1[:, 0], t1[:, 1], label='Array 1 Traj 0', color='blue')
        plt.scatter(t2[0, 0], t2[0, 1], label='Array 2 Traj 0', color='orange')
        # plt.plot(t2[:, 0], t2[:, 1], label='Array 2 Traj 0', color='orange')
        plt.scatter(t3[0, 0], t3[0, 1], label='Array 3 Traj 0', color='green')
        # plt.plot(t3[:, 0], t3[:, 1], label='Array 3 Traj 0', color='green')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('First Trajectory from Each Array')
    plt.grid(True)
    plt.show()

# Example usage
plot_all(expert_data_1, expert_data_2, expert_data_3)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Palatino Linotype'] + plt.rcParams['font.serif']

def plot_expert(arr1, arr2, arr3):
    """
    Plots the paths of three agents.
    - Earlier trajectories muted.
    - Final trio emphasized with bold, thicker lines.
    """
    plt.figure(figsize=(8, 6))
    ax = plt.gca()

    # Define colors (normal vs muted)
    colors = {
        "Agent 1": ("blue", "lightblue"),
        "Agent 2": ("orange", "navajowhite"),
        "Agent 3": ("green", "lightgreen"),
    }

    highlight = 10

    for i in range(len(arr1)):
        if i != highlight:
            # Earlier: muted thin lines
            plt.plot(arr1[i][:, 0], arr1[i][:, 1], color=colors["Agent 1"][1], alpha=0.4, linewidth=1)
            plt.plot(arr2[i][:, 0], arr2[i][:, 1], color=colors["Agent 2"][1], alpha=0.4, linewidth=1)
            plt.plot(arr3[i][:, 0], arr3[i][:, 1], color=colors["Agent 3"][1], alpha=0.4, linewidth=1)

    plt.plot(arr1[highlight][:, 0], arr1[highlight][:, 1], color=colors["Agent 1"][0], linewidth=3, label="Agent 1")
    plt.plot(arr2[highlight][:, 0], arr2[highlight][:, 1], color=colors["Agent 2"][0], linewidth=3, label="Agent 2")
    plt.plot(arr3[highlight][:, 0], arr3[highlight][:, 1], color=colors["Agent 3"][0], linewidth=3, label="Agent 3")

    # Labels and title with larger fonts
    plt.xlabel("X-Position", fontsize=16)
    plt.ylabel("Y-Position", fontsize=16)
    plt.title("Three-Agent Road Crossing", fontsize=18)

    # Legend formatting
    plt.legend(frameon=False, fontsize=14)

    # Remove grid and extra spines
    plt.grid(False)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    plt.show()

# Example call
plot_expert(expert_data_1, expert_data_2, expert_data_3)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams
from matplotlib import font_manager as fm

# --- Load your custom font ---
font_path = "utils/palr45w.ttf"  # << replace with actual path
custom_font = fm.FontProperties(fname=font_path)

# Set rcParams globally to use this font
rcParams['font.family'] = custom_font.get_name()

def plot_expert(arr1, arr2, arr3):
    """
    Plots the paths of three agents.
    - Earlier trajectories muted.
    - Final trio emphasized with bold, thicker lines and arrows.
    """
    plt.figure(figsize=(8, 6))
    ax = plt.gca()

    # Define colors (normal vs muted)
    colors = {
        "Agent 1": ("blue", "lightblue"),
        "Agent 2": ("orange", "navajowhite"),
        "Agent 3": ("green", "lightgreen"),
    }

    highlight = 10  # index of highlighted trajectory

    # Muted earlier trajectories
    for i in range(len(arr1)):
        if i != highlight:
            ax.plot(arr1[i][:, 0], arr1[i][:, 1], color=colors["Agent 1"][1], alpha=0.4, linewidth=1)
            ax.plot(arr2[i][:, 0], arr2[i][:, 1], color=colors["Agent 2"][1], alpha=0.4, linewidth=1)
            ax.plot(arr3[i][:, 0], arr3[i][:, 1], color=colors["Agent 3"][1], alpha=0.4, linewidth=1)

    # Bold highlighted trajectories
    ax.plot(arr1[highlight][:, 0], arr1[highlight][:, 1], color=colors["Agent 1"][0], linewidth=3, label="Agent 1")
    ax.plot(arr2[highlight][:, 0], arr2[highlight][:, 1], color=colors["Agent 2"][0], linewidth=3, label="Agent 2")
    ax.plot(arr3[highlight][:, 0], arr3[highlight][:, 1], color=colors["Agent 3"][0], linewidth=3, label="Agent 3")

    # ---- Add arrows to show direction ----
    def add_arrow(traj, color):
        n = len(traj)
        mid = n // 4
        for k in [1, 2, 3]:
            ax.annotate(
                "",
                xy=traj[k*mid + 1],
                xytext=traj[k*mid - 1],
                arrowprops=dict(arrowstyle="->", color=color, lw=5),
            )

    add_arrow(arr1[highlight], colors["Agent 1"][0])
    add_arrow(arr2[highlight], colors["Agent 2"][0])
    add_arrow(arr3[highlight], colors["Agent 3"][0])

    # Labels and title with larger fonts
    ax.set_xlabel("X-Position", fontsize=16)
    ax.set_ylabel("Y-Position", fontsize=16)
    ax.set_title("Three-Agent Road Crossing", fontsize=18)

    # Legend formatting
    ax.legend(frameon=False, fontsize=14)

    # Remove grid and extra spines
    ax.grid(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    plt.show()

# Example call
plot_expert(expert_data_1, expert_data_2, expert_data_3)
```
